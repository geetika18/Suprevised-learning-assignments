```python
# input array
in_array = np.array([1,2,3])
exp_array = np.exp(in_array) #using numpy function exp() to calculate e^z

print("Input: ", in_array)
print("Output: ", exp_array)
```

```python
#implementing sigmoid function
def sigmoid(z):
    """
    compute sigmoid of z
    args: 
     z(ndarray): a scalar, numpy array of any size
    returns:
     g(ndarray): sigmoid(z), with the same shape as z

    """
    g = 1/(1+np.exp(-z))
    return g
```

```python
#check the sigmoid function output

z_tmp = np.arange(-10,11)
#print(z)
y = sigmoid(z_tmp)

print("  Input(z), Output(sigmoid(z))")
print(np.c_[z_tmp, y])
```

```python
#plot z vs sigmoid(z)
fig, ax = plt.subplots(1,1,figsize =(5,3))
ax.plot(z_tmp, y, c='b')
ax.set_title('Z vs Sigmoid(z)')
ax.set_ylabel('sigmoid(z)')
ax.set_xlabel('z')
draw_vthresh(ax,0)
#plt.show()
```
